```python
import matplotlib.pyplot as plt
import powerlaw
from scipy.interpolate import make_interp_spline, BSpline

from load_graph import *

import seaborn as sns
pal = sns.color_palette("colorblind")
colors = [pal[1], "#2b8cbe" ,pal[2]]
dpi = 800
```

```python
## DBLP graph
_, real = load_snap("data/com-dblp.ungraph.txt", "data/com-dblp.all.cmty.txt")
abcdoo, _ = load_coms("data/abcdoo_dblp_d2_com.dat", True)
cbk, _ = load_coms("data/cbk_dblp.dat", True)

print(f"{len(real)} real communities.")
print(f"{len(abcdoo)} abcdoo communities.")
print(f"{len(cbk)} cbk communities.")

options = [real, abcdoo, cbk]
names = ["DBLP", r"ABCD+o$^2$", "CBK"]

fig, ax = plt.subplots(dpi=dpi)
for coms, name, color in zip(options, names, colors):
    X = powerlaw.Fit([len(x) for x in coms], xmin=1) # actual fit does not matter
    X.plot_ccdf(ax=ax, linewidth=3, original_data=False, linestyle="solid", label=name, color=color)

ax.set_title("DBLP Community Size Distribution", fontsize=16)
ax.legend(fontsize=13)
ax.set_xlabel("Community Size", fontsize=15)
ax.set_ylabel("Complementary CDF", fontsize=15)
ax.tick_params(labelsize=12)
```
